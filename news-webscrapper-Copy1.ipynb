```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np

import slugify
from slugify import slugify

url = 'https://www.dovislex.eu/en/index.php?page=news'
content = requests.get(url)

soup = BeautifulSoup(content.text, 'html.parser')

title = soup.find_all('h3')
title_list = []
for i in title:
    n = str(i)
    title_list.append(n)

    
article = soup.find_all('div', class_="clearfix")          
articles = []

for j in range(0, 36, 1):
    element = str(article[j].find_all('ul'))
    element = element[1:-1]
    articles.append(element)
    
date = soup.find_all('h4')
date = date[1:]


date_e = []
for i in range(0, len(date)):
    date_e.append(str(date[i])) 


data = {}
id_range = range(1, len(soup.find_all('h3'))+1, 1)

id_ = list(id_range)
for i in range(0, len(soup.find_all('h3')), 1):
    data[i] = {'id': id_[i],
               'user_id' : 1,
               'date' : date_e[i],
               'title': title_list[i],
               'slug' : title_list[i],
               'view' : 0,
               'image' : '',
               'body': articles[i],
               'published' : 1
               }


df = pd.DataFrame(data)
df = df.transpose()

for i in range(0,len(date)):
    df['date'][i] = df['date'][i].replace('<h4> ', '').replace('</h4>', '').replace('<h4>', '').strip()
    
for i in range(0,len(df['title'])):
    df.slug[i] = df.slug[i].split('>',)[1].split('<',)[0]
    
for i in range(0,len(df['slug'])):
        df['slug'][i] = slugify(df['slug'][i], replace_latin=True)
        
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", None)

print(df)

df.to_csv(r'/Users/bi/Documents/webscraper2.csv', index=False)

```
